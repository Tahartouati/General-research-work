```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/dayTimeModerate.csv')
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df.head()
```

```python
correlationVector = df.corr().iloc[[3]]
```

```python
correlationVector = correlationVector.drop(['SEQN','moderate'],axis=1)
```

```python
correlationVector
```

```python
sortedActiveCorr = correlationVector.abs().unstack().sort_values(ascending=False)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
  print(sortedActiveCorr)
```
