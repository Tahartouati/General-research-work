```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/dayTimeClassDF.csv')
```

```python
df.drop(['Unnamed: 0'],axis = 1,inplace = True)
```

```python
df.shape
```

```python
df.head()
```

```python
categories = []
for index, row in df.iterrows():
    if row['CAT'] == 'intense':
            categories.append(4)
    elif row['CAT'] == 'moderate':
            categories.append(3)
    elif row['CAT'] == 'light':
            categories.append(2)
    else:
            categories.append(1)
```

```python
df['ordCAT'] = categories
```

```python
df.shape
```

```python
corrVec = df.corr().iloc[[94]].drop(['SEQN','intense','moderate','light','sedentary'],axis = 1)
```

```python
sortedActiveCorr = corrVec.abs().unstack().sort_values(ascending=False)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
  print(sortedActiveCorr)
```
