### active people: intense = mean+ 4std

```python
dfmean = df[['intense']].mean() 
dfstd = df[['intense']].std()
val = dfmean +4*dfstd
val
```

```python
active = df.loc[df['intense'] > 20]
active.shape
```

```python
activeCorr = active.corr().iloc[[1]]
activeCorr
```

```python
sortedActiveCorr = activeCorr.abs().unstack().sort_values(ascending=False)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
  print(sortedActiveCorr)
```
