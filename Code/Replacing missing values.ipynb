```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('dataset_V4.csv')
categorical_var = [
 'HUQ010',
 'HUQ050',
 'HUQ060',
 'DPQ010',
 'DPQ030',
 'DPQ040',
 'DPQ050','DPQ080']
numerical_var = [
 'PAD680',
 'HUD080',
 'HSQ470',
 'HSQ490',
 'HSQ493',
 'HSQ496',
 'MCQ180A',
 'MCQ180B',
 'MCQ180C',
 'MCQ180D',
 'MCQ180E','MCQ180F','MCQ180K']
```

```python
dataset = data[['PFQ061B','PFQ061C','PFQ061H']+numerical_var+categorical_var]
```

```python
correlation = dataset.corr().iloc[[0,1,2]].drop(['PFQ061B','PFQ061C','PFQ061H'],axis=1)
```

```python
for column in numerical_var:
    dataset[column].replace(np.nan,dataset[column].mean(),inplace = True)
```

```python
for column in categorical_var:
    dataset[column].replace(np.nan,float(dataset[column].mode()),inplace = True)
```

```python
XY1 = dataset.drop(['PFQ061C','PFQ061H'],axis = 1).dropna(subset = ["PFQ061B"])
XY2 = dataset.drop (['PFQ061B','PFQ061H'], axis = 1).dropna(subset = ["PFQ061C"])
XY3 = dataset.drop (['PFQ061B','PFQ061C'], axis = 1).dropna(subset = ["PFQ061H"])
```

```python
XY1 = dataset.drop(['PFQ061C','PFQ061H'],axis = 1)
XY1.dropna(subset = ["PFQ061B"], inplace = True)
```

```python
XY1.to_csv('PFQ061B.csv')
XY2.to_csv('PFQ061C.csv')
XY3.to_csv('PFQ061H.csv')
```
