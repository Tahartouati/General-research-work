```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("dataset_V1.csv") 
```

```python
data.head()
```

```python
data.shape
```

```python
column_7 = ['PFQ061B','PFQ061C','PFQ061H','HUQ010','HUQ020','HUQ060','DPQ010','DPQ020','DPQ030','DPQ040','DPQ050','DPQ060','DPQ070','DPQ080','DPQ090','DPQ100','CIQD002','CIQD003','CIQD004','CIQD005']
```

```python
column_77 = ['PAQ670','PAQ706','PAQ710','PAQ715','HUQ050','HSQ470','HSQ480','HSQ490','HSQ493','HSQ496','HSQ496']
```

```python
column_7777 = ['PAD675','PAD680']
```

```python
column_77777 = ['HUD080','MCQ180B','MCQ180C','MCQ180D','MCQ180E','MCQ180F','MCQ180M','MCQ180K']
```

```python
data.shape
```

```python
for column in column_7 :
    data.drop(data.index[data[column] == 7], inplace = True)
    data.drop(data.index[data[column] == 9], inplace = True)
for column in column_77 :
    data.drop(data.index[data[column] == 77], inplace = True)
    data.drop(data.index[data[column] == 99], inplace = True)
for column in column_7777 :
    data.drop(data.index[data[column] == 7777], inplace = True)
    data.drop(data.index[data[column] == 9999], inplace = True)
for column in column_77777 :
    data.drop(data.index[data[column] == 77777], inplace = True)
    data.drop(data.index[data[column] == 99999], inplace = True)
```

```python
data.shape
```

```python
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
data.to_csv('dataset_V2.csv')
```
